## 图像格式

图像格式是指用于存储和传输图像数据的特定文件格式。不同的图像格式有不同的特点和用途。以下是一些常见的图像格式及其特点：

1. **JPEG（Joint Photographic Experts Group）**：JPEG 是一种广泛用于图像压缩的格式。它通过基于人眼视觉特性的压缩算法，减小图像文件的大小，但会引入一定程度的失真。JPEG 格式适合于存储照片和彩色图像。
2. **PNG（Portable Network Graphics）**：PNG 是一种**无损压缩**的图像格式，适合存储需要保留细节和透明度的图像。相比于 JPEG，PNG 文件通常更大，但不会引入压缩失真。它常用于存储图标、图形和带有透明背景的图像。
3. **GIF（Graphics Interchange Format）**：GIF 是一种支持动画和透明度的图像格式。它使用 LZW 压缩算法，适用于简单动画、图标和图形。
4. **BMP（Bitmap）**：BMP 是一种**无损的**位图图像格式，它以像素的颜色信息直接存储图像数据。BMP 文件通常较大，但可以保留图像的高质量细节。
5. **TIFF（Tagged Image File Format）**：TIFF 是一种通用的**无损压缩**或无压缩的图像格式，适合存储高质量的图像数据。它支持多种颜色模型和深度，并且可以存储多帧图像。

### 颜色空间变换

````python
import matplotlib.pyplot as plt
from PIL import ImageEnhance
from PIL import Image

# 调整图像亮度
def adjust_brightness(image, factor):
    enhancer = ImageEnhance.Brightness(image)
    return enhancer.enhance(factor)

# 调整图像对比度
def adjust_contrast(image, factor):
    enhancer = ImageEnhance.Contrast(image)
    return enhancer.enhance(factor)

# 调整图像饱和度
def adjust_saturation(image, factor):
    enhancer = ImageEnhance.Color(image)
    return enhancer.enhance(factor)

# 示例用法
image = Image.open("琨琨.png")

# 调整亮度为原来的2倍
brightened_image = adjust_brightness(image, 2.0)

# 调整对比度为原来的0.5倍
contrasted_image = adjust_contrast(image, 0.5)

# 调整饱和度为原来的1.5倍
saturated_image = adjust_saturation(image, 1.5)

# 创建2x2的子图
fig, axs = plt.subplots(2, 2, figsize=(20, 15))

# 显示原图
axs[0, 0].imshow(image)
axs[0, 0].set_title('Original')

# 显示亮度调整后的图像
axs[0, 1].imshow(brightened_image)
axs[0, 1].set_title('Brightened')

# 显示对比度调整后的图像
axs[1, 0].imshow(contrasted_image)
axs[1, 0].set_title('Contrasted')

# 显示饱和度调整后的图像
axs[1, 1].imshow(saturated_image)
axs[1, 1].set_title('Saturated')

# 移除子图的坐标轴
for ax in axs.flat:
    ax.axis('off')

# 调整子图之间的间距
plt.tight_layout()

# 显示结果
plt.show()
````

### 滤波

滤波是一种常见的图像处理技术，用于平滑图像、去除噪声或改变图像的频谱特性。滤波器通常是一个二维的小窗口或卷积核，通过对图像中的像素进行加权平均或加权求和来改变图像的外观。常见的滤波方法包括：
- **均值滤波**：将每个像素的值替换为其邻域像素的平均值，可以平滑图像并减少噪声。
- **高斯滤波**：使用加权平均值，其中每个像素的权重是由高斯函数计算的。高斯滤波可以更好地保留图像细节同时平滑噪声。
- **中值滤波**：将每个像素的值替换为其邻域像素值的中值。中值滤波能够有效地去除图像中的椒盐噪声等异常值。

````python
import cv2

# 读取图像
image = cv2.imread('琨琨.png')
# 均值滤波
image_mean = cv2.blur(image, (10, 10))
# 高斯滤波
image_gaussian = cv2.GaussianBlur(image, (5, 5), 3, 3)
# 中值滤波
image_median = cv2.medianBlur(image, 9)
````

cv2.GaussianBlur() 是 OpenCV 库中的一个函数，用于应用高斯滤波（Gaussian Blur）来平滑图像。

函数的语法如下：

```python
blurred_image = cv2.GaussianBlur(src, ksize, sigmaX[, dst[, sigmaY[, borderType]]])
```
参数说明：

- src：要进行滤波的输入图像。
- ksize：滤波器的内核大小。它是一个正的奇数，例如 (5, 5)、(9, 9) 等。内核大小决定了高斯滤波器的模糊程度。
- sigmaX：高斯内核在 X 方向上的标准差。
- dst（可选）：输出图像，与输入图像具有相同的尺寸和深度。
- sigmaY（可选）：高斯内核在 Y 方向上的标准差。如果为零，则使用 sigmaX 的值。如果同时 sigmaX 和 sigmaY 均为零，则通过 ksize.width 和 ksize.height 计算得到的值分别用于 sigmaX 和 sigmaY。
borderType（可选）：用于推断图像边界的像素填充方式，具体取值可参考 OpenCV 文档中的 borderType。

### 边缘检测
边缘检测是一种用于寻找图像中不同区域之间边界的技术。边缘通常表示图像中像素强度发生明显变化的地方。常用的边缘检测算法包括：
- **Sobel算子**：通过计算图像中每个像素的梯度，检测出边缘的强度和方向。
- **Prewitt算子**：类似于Sobel算子，通过计算图像中每个像素的梯度来检测边缘。
- **Canny边缘检测**：Canny算法结合了多个步骤，包括高斯滤波、梯度计算、非极大值抑制和双阈值处理，以检测出准确的边缘。

````python
import cv2
import matplotlib.pyplot as plt

# 读取图像
image = cv2.imread("琨琨.png")

# 转换图像为灰度图
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Sobel算子边缘检测
sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
sobel_magnitude = cv2.magnitude(sobel_x, sobel_y)

# Prewitt算子边缘检测
prewitt_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
prewitt_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
prewitt_magnitude = cv2.magnitude(prewitt_x, prewitt_y)

# Canny边缘检测
canny_edges = cv2.Canny(gray_image, 100, 200)

# 创建2x2的子图
fig, axs = plt.subplots(2, 2)

# 显示原图
axs[0, 0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('Original')

# 显示Sobel算子边缘检测结果
axs[0, 1].imshow(sobel_magnitude, cmap='gray')
axs[0, 1].set_title('Sobel')

# 显示Prewitt算子边缘检测结果
axs[1, 0].imshow(prewitt_magnitude, cmap='gray')
axs[1, 0].set_title('Prewitt')

# 显示Canny边缘检测结果
axs[1, 1].imshow(canny_edges, cmap='gray')
axs[1, 1].set_title('Canny')

# 移除子图的坐标轴
for ax in axs.flat:
    ax.axis('off')

# 调整子图之间的间距
plt.tight_layout()

# 显示结果
plt.show()
````

### 特征提取
特征提取是从图像中提取有区分性的信息以描述图像内容的过程。这些特征可以代表图像的某些视觉属性，如颜色、纹理、形状等。常见的特征包括：
- **颜色直方图**：计算图像中各种颜色的分布情况，可以用于图像分类、图像搜索等任务。
- **纹理特征**：用于描述图像的纹理特性，比如灰度共生矩阵（GLCM）和小波变换等。
- **形状特征**：描述图像中物体的形状和结构，例如边界。

````python
import cv2
import matplotlib.pyplot as plt
from skimage.feature import greycomatrix, greycoprops
from skimage.measure import find_contours
import numpy as np

# 读取图像
image = cv2.imread("琨琨.png")

# 转换图像为灰度图
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 计算颜色直方图
histogram = cv2.calcHist([image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])

# 计算灰度共生矩阵
glcm = greycomatrix(gray_image, [1], [0, np.pi/4, np.pi/2, 3*np.pi/4], levels=256, symmetric=True, normed=True)

# 计算纹理特征（对比度、能量、相关性、均匀性）
contrast = greycoprops(glcm, 'contrast')
energy = greycoprops(glcm, 'energy')
correlation = greycoprops(glcm, 'correlation')
homogeneity = greycoprops(glcm, 'homogeneity')

# 寻找图像边界
contours = find_contours(gray_image, 0.8)

# 创建2x2的子图
fig, axs = plt.subplots(2, 2)

# 显示原图
axs[0, 0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('Original')

# 显示颜色直方图
axs[0, 1].bar(range(len(histogram.flatten())), histogram.flatten())
axs[0, 1].set_title('Color Histogram')

# 显示纹理特征
axs[1, 0].bar(range(len(contrast.flatten())), contrast.flatten())
axs[1, 0].set_title('Texture Features')

# 显示边界
axs[1, 1].imshow(gray_image, cmap='gray')
for contour in contours:
    axs[1, 1].plot(contour[:, 1], contour[:, 0], linewidth=2)
axs[1, 1].set_title('Contours')

# 移除子图的坐标轴
for ax in axs.flat:
    ax.axis('off')

# 调整子图之间的间距
plt.tight_layout()

# 显示结果
plt.show()
````
